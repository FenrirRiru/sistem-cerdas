```python
import os
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50V2
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
dataset_path = '/content/drive/MyDrive/dataset/archive(1)/Diagnosis_of_Diabetic_Retinopathy/valid/'

datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

train_generator = datagen.flow_from_directory(
    dataset_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

val_generator = datagen.flow_from_directory(
    dataset_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)

#debug
print("Class Indices (Label mapping):")
print(train_generator.class_indices)
```

```python
base_model = ResNet50V2(
    include_top=False,
    input_shape=(224, 224, 3),
    weights='imagenet'
)
base_model.trainable = False  # freeze model agar tidak belajar ulang

model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(1, activation='sigmoid')  # klasifikasi biner
])

model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
checkpoint = ModelCheckpoint(
    '/content/drive/MyDrive/Model/best_model.h5',
    monitor='val_accuracy',
    save_best_only=True,
    verbose=1
)

earlystop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)
```

```python
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10,
    callbacks=[checkpoint, earlystop]
)
```

```python
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.legend()
plt.title('Akurasi')
plt.show()

plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.legend()
plt.title('Loss')
plt.show()
```

```python
import os
import shutil
import numpy as np
from tensorflow.keras.preprocessing import image

uji_path = '/content/drive/MyDrive/dataset/archive(2)/DRD_Dataset'  # path gambar uji
output_file = '/content/drive/MyDrive/dataset/output/hasil_deteksi.csv'

# Folder tujuan salinan
folder_terkena = '/content/drive/MyDrive/dataset/output/terkena'
folder_tidak_terkena = '/content/drive/MyDrive/dataset/output/tidak_terkena'

# Buat folder jika belum ada
os.makedirs(folder_terkena, exist_ok=True)
os.makedirs(folder_tidak_terkena, exist_ok=True)

# Mulai proses prediksi dan pemisahan gambar
with open(output_file, 'w') as f:
    f.write('Nama File,Prediksi,Confidence\n')

    for fname in os.listdir(uji_path):
        if fname.lower().endswith(('.jpg', '.jpeg', '.png')):
            path = os.path.join(uji_path, fname)
            try:
                img = image.load_img(path, target_size=(224, 224))
                x = np.expand_dims(image.img_to_array(img) / 255.0, axis=0)

                pred = model.predict(x, verbose=0)[0][0]
                label = 'terkena' if pred > 0.5 else 'tidak terkena'
                f.write(f'{fname},{label},{pred:.2f}\n')

                # Salin gambar ke folder yang sesuai
                if label == 'terkena':
                    shutil.copy(path, os.path.join(folder_terkena, fname))
                else:
                    shutil.copy(path, os.path.join(folder_tidak_terkena, fname))

            except Exception as e:
                print(f" Gagal memproses {fname}: {e}")
```
